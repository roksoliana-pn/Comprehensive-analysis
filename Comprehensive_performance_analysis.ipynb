For better data perception columns names should be renamed.

```python
# Changing columns order
df = df[["order_id", "order_date", "shipping_date", "processing_duration", 'order_priority',
       "product_id", "category", "units_sold", "unit_price", "unit_cost", "sales_channel",
       "country_name", "region", "sub-region"]]
df.head()
```

To analyze the data better and to improve data display the order of columns was changed.

##Basic analysis

```python
# Calculation of Cost, Revenue and Profit
df["revenue"] = df["unit_price"] * df["units_sold"]
df["cost"] = df["unit_cost"] * df["units_sold"]
df["profit"] = df["revenue"] - df["cost"]
df.head()
```

The columns with calculation of Revenue, Cost and Profit were added to the dataset.

```python
# Calculation of total metrics
total_orders_number = df["order_id"].count()
total_quantity_sold = df["units_sold"].sum()
total_revenue = df["revenue"].sum()
total_cost = df["cost"].sum()
total_profit = df["profit"].sum()
number_of_countries = df["country_name"].nunique()
print(f"Main metrics:\nTotal number of orders: {total_orders_number};\n- Total number of products sold: {total_quantity_sold};\n- Total revenue: {total_revenue} $;\n- Total cost: {total_cost} $;\n- Total profit: {total_profit} $;\n- Number of countries where orders were placed: {number_of_countries}.")
```

```python
# Main fianncial indicators dynamics

df["year"] = df["order_date"].dt.year

df_metrics = pd.melt(df, id_vars="year", value_vars=["revenue", "cost", "profit"],
                  var_name='Metric', value_name='value')
sns.lineplot(data=df_metrics, x="year", y="value", hue="Metric", palette="mako")

plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x):,}'))

plt.title("Main financial metrics dynamics", fontweight="bold")
plt.xlabel("Year", fontweight="bold")
plt.ylabel("Financial result, $", fontweight="bold")

plt.show()
```

The dynamics of key financial indicators was assessed, which revealed that the company's highest results date back to 2012, and in the period from 2014 to 2017 there was a gradual decline in indicators, which are now at the lowest level for the entire period under study.

#Revenue dynamics analysis according to the days of the week

```python
# Days of the week

df["day_name"] = df["order_date"].dt.day_name()
day_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
df["day_name"] = pd.Categorical(df["day_name"], categories=day_order, ordered=True)

days_revenue = pd.pivot_table(df, index="day_name", values="revenue", aggfunc="sum")
days_revenue
```

```python
# Vizualization of revenue by days of the week and category

fig, axes = plt.subplots(2, 1, figsize=(14, 8))

revenue_by_day = df.groupby("day_name")["revenue"].sum().reindex(day_order)

sns.barplot(x=revenue_by_day.index, y=revenue_by_day.values, palette="magma",
            ax=axes[0])
axes[0].yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
axes[0].set_title("Total Revenue by Day of the Week", fontweight="bold")
axes[0].set_xlabel("Day of the Week", fontweight="bold")
axes[0].set_ylabel("Revenue, $", fontweight="bold")
axes[0].set_ylim(0, 300_000_000)

for i, value in enumerate(revenue_by_day.values):
    axes[0].text(i, value + 5_000_000, f'{int(value):,}', ha='center', va='bottom', fontsize=9)

sns.barplot(x="category", y="revenue", hue="day_name", data=df,
            errorbar=("ci", False), palette="magma", ax=axes[1])
axes[1].yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
axes[1].set_title("Revenue by Category and Day of the Week", fontweight="bold")
axes[1].set_xlabel("Category", fontweight="bold")
axes[1].set_ylabel("Revenue, $", fontweight="bold")
axes[1].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```

The next stage of the analysis was to estimate the level of revenue depending on the day of the week. As a result, we can conclude that most orders were placed on Monday, Wednesday, and Friday, while the lowest number of orders were placed on Tuesday. However, it is important to understand that the difference in the distribution is only 1-2%, so it is fairly even.

Analyzing the total revenue by category and day of the week reveals that Household and Office Supplies generate the highest sales, particularly on Mondays, while Beverages and Fruits have the lowest overall revenue. Revenue trends differ significantly across product types; for example, Household and Office Supplies peak early in the week, whereas Meat and Cosmetics see a boost towards the weekend. Notably, a mid-week dip in revenue is observed across several categories before a weekend recovery in some.
